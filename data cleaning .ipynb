```python
import pandas as pd

df = pd.read_csv("trials_1000_clean (1).csv")
print(df.shape)
df.head()
```

```python
df = df.rename(columns={
    "nctId": "nct_id",
    "briefTitle": "brief_title",
    "officialTitle": "official_title",
    "conditions": "condition",
    "sex": "gender",
    "eligibilityCriteria": "eligibility_text",
    "minimumAge": "age_min_raw",
    "maximumAge": "age_max_raw",
    "studyType": "study_type",
    "overallStatus": "overall_status"
})
df.head()
```

```python
df['age_min'] = df['age_min_raw'].str.extract('(\d+)').astype(float)
df['age_max'] = df['age_max_raw'].str.extract('(\d+)').astype(float)
print(df[['age_min_raw', 'age_min', 'age_max_raw', 'age_max']].head(10))
df=df.drop(columns=['age_min_raw', 'age_max_raw'])
print(df.shape)
```

```python
df['age_min'].fillna(df['age_min'].median(), inplace=True)
df['age_max'].fillna(df['age_max'].median(), inplace=True)
df = df.dropna(subset=['official_title'])
df = df[df['age_min'] <= df['age_max']]
df = df.drop_duplicates()
print(f"Data shape after cleaning: {df.shape}")
df.head()
```

```python
df['condition'] = df['condition'].str.lower()
df['conditions_list'] = df['condition'].str.split(';').apply(lambda x: [c.strip() for c in x])
print(df[['condition', 'conditions_list']].head(10))
df.drop(columns=['condition'], inplace=True)
print(df.shape)
```

```python
import re

def clean_eligibility(text):
    if pd.isna(text):
        return ""
    text = re.sub(r"<[^>]+>", " ", text)   # remove HTML
    text = re.sub(r"\s+", " ", text)       # normalize whitespace
    return text.strip()
```

```python
df.to_csv("cleaned_trials.csv", index=False)
print(f"Final cleaned trial count: {len(df)}")
print(f"Shape of cleaned DataFrame: {df.shape}")
```
